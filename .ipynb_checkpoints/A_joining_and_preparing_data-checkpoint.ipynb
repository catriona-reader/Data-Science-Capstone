<b style="color:Tomato;">Cleaning aggregate data</b>

The aggregate census data came in multiple csv files, each with it's own data description file. I start by reading the csvs into dataframes, choosing which variables to keep as predictors and renaming the columns for ease of reading. Finally the separate dataframes are merged together on 'GeographyCode'. 

I've commented out all but the first data description file, so that the reader can get a sense of the format without getting stuck in the detail. 

```python
import numpy as np
import pandas as pd 
```

```python
path1 = 'datasets/aggregate_data/KS101EWDATA.CSV'
p1_desc = 'datasets/aggregate_data/KS101EWDESC0.CSV'
df1 = pd.read_csv(path1)
df1_desc = pd.read_csv(p1_desc)
```

```python
print('Shape: ', df1.shape)
df1.head(2)
```

```python
df1_desc
```

```python
cols_to_drop = ['KS101EW0002', 'KS101EW0003', 'KS101EW0004', 'KS101EW0005', 'KS101EW0006']
df1 = df1.drop(columns=cols_to_drop)

rename_columns = ['GeographyCode', 'Tot_people', 'Tot_area_hect', 
                  'Density', 'Male%', 'Female%', 'Household_living%', 'Communal_living%']

df1.columns = rename_columns
df1.head(2)
```

```python
path2 = 'datasets/aggregate_data/KS104EWDATA.CSV'
p2_desc = 'datasets/aggregate_data/KS104EWDESC0.CSV'
df2 = pd.read_csv(path2)
df2_desc = pd.read_csv(p2_desc)
```

```python
df2.head(2)
```

```python
#df2_desc
```

```python
#simply keeping living in a couple or not - which I'll show with total percentage in a couple
#will add these columns together once I've converted all values to float
cols = ['GeographyCode','KS104EW0009', 'KS104EW0010']
```

```python
df2 = df2[cols]
rename_columns = ['GeographyCode', 'couple1', 'couple2']
df2.columns = rename_columns
print('Shape: ', df2.shape)
df2.head(2)
```

```python
path3 = 'datasets/aggregate_data/KS301EWDATA.CSV'
p3_desc = 'datasets/aggregate_data/KS301EWDESC0.CSV'
df3 = pd.read_csv(path3)
df3_desc = pd.read_csv(p3_desc)
```

```python
#df3_desc
```

```python
print('Shape: ', df3.shape)
df3.head(2)
```

```python
cols = ['GeographyCode','KS301EW0023', 'KS301EW0024', 'KS301EW0025', 'KS301EW0026', 
        'KS301EW0027']

df3 = df3[cols]

rename_columns = ['GeographyCode', 'Health_vg', 'Health_g', 'Health_f', 'Health_b','Health_vb']
df3.columns = rename_columns
df3.head(2)
```

```python
path4 = 'datasets/aggregate_data/KS501EWDATA.CSV'
p4_desc = 'datasets/aggregate_data/KS501EWDESC0.CSV'
df4 = pd.read_csv(path4)
df4_desc = pd.read_csv(p4_desc)
```

```python
print('Shape', df4.shape)
df4.head(2)
```

```python
#df4_desc
```

```python
cols = ['GeographyCode','KS501EW0014', 'KS501EW0015', 'KS501EW0016', 'KS501EW0017', 
        'KS501EW0018','KS501EW0019', 'KS501EW0020']
df4 = df4[cols]

rename_columns = ['GeographyCode','Qual_0', 'Qual_1', 'Qual_2', 
                  'Qual_A', 'Qual_3', 'Qual_4', 'Qual_Other']
df4.columns = rename_columns
df4.head(2)
```

<i style="color:Tomato;">Next step: merging the separate dataframes together</i>

```python
data = pd.merge(df1, df2, how='outer', on='GeographyCode')
```

```python
data = pd.merge(data, df3, how='outer', on='GeographyCode')
```

```python
data = pd.merge(data, df4, how='outer', on='GeographyCode')
```

```python
print(data.shape)
data.head(2)
```

```python
data.isna().sum()
```

<i style="color:Tomato;">The data has been read in as strings - I'll need to convert everything bar Geography code to a float value</i>

```python
geo = data.pop('GeographyCode')
```

```python
def float_converter(x):
    try:
        return float(x)
    except:
        return np.nan
```

```python
data = data.applymap(float_converter)
```

```python
data.insert(0,'GeographyCode', geo)
```

```python
data.head(2)
```

```python
#testing to make sure that's worked
data.Tot_people[0].dtype
```

```python
#did it result in any NaN values? 
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
#Now doing some column groupings as the values are floats 
#These both indicate instances of living in a couple, so I'll combine them 
data['couple_living'] = data.couple1 + data.couple2
```

```python
data.drop(columns=['couple1', 'couple2'], inplace=True)
```

```python
data.describe()
```

<i style="color:Tomato;">I'm happy that this data is clean, so I'm writing it a csv file</i>

```python
export_csv = data.to_csv(r'datasets/clean_data/census.csv', index = None, header=True)
```
